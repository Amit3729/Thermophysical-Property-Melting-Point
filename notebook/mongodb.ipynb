```python
import pandas as pd
from pymongo import MongoClient
import os
```

```python
df1 = pd.read_csv('/Users/spkunwar/Desktop/Thermophysical-Property-Melting-Point/melting-point/test.csv')
df2 = pd.read_csv('/Users/spkunwar/Desktop/Thermophysical-Property-Melting-Point/melting-point/train.csv')
print(df1.head())
print(df2.head())
```

```python
from dotenv import load_dotenv

load_dotenv()

data1 = df1.to_dict(orient='records')
data2 = df2.to_dict(orient='records')

connection_uri = os.getenv('MONGODB_URI')

if not connection_uri:
    raise ValueError("Mongobd not found in env")

client = MongoClient(connection_uri) #connect to mongodb
db = client[os.getenv('DB_NAME')]
collection = db[os.getenv('COLLECTION_NAME')] 

# Separate collections for each CSV
collection1 = db["csv1_data"]
collection2 = db["csv2_data"]

# Insert data
collection1.insert_many(data1)
collection2.insert_many(data2)
```
